## üß™ –ó–∞–¥–∞–Ω–∏–µ: AutoML –Ω–∞ sklearn-–¥–∞—Ç–∞—Å–µ—Ç–µ —Å FLAML

```python
from flaml import AutoML
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞—Ç–∞—Å–µ—Ç
X, y = load_wine(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# –ó–∞–ø—É—Å–∫–∞–µ–º AutoML
automl = AutoML()
automl.fit(X_train=X_train, y_train=y_train, task="classification", time_budget=60)

# –û—Ü–µ–Ω–∫–∞
y_pred = automl.predict(X_test)
acc_automl = accuracy_score(y_test, y_pred)
print("FLAML Accuracy:", acc_automl)
```

## üîç –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å —Ä—É—á–Ω–æ–π –º–æ–¥–µ–ª—å—é (RandomForest)

```python

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

acc_rf = accuracy_score(y_test, rf_pred)
print("Random Forest Accuracy:", acc_rf)
```
