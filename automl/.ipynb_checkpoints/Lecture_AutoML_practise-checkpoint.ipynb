```python
## âœ… Auto-sklearn: ÐšÐ»Ð°ÑÑÐ¸Ñ„Ð¸ÐºÐ°Ñ†Ð¸Ñ Breast Cancer
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import autosklearn.classification
```

```python
# Ð—Ð°Ð³Ñ€ÑƒÐ·ÐºÐ° Ð¸ Ñ€Ð°Ð·Ð±Ð¸ÐµÐ½Ð¸Ðµ Ð´Ð°Ð½Ð½Ñ‹Ñ…
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# ÐžÐ±ÑƒÑ‡ÐµÐ½Ð¸Ðµ AutoML Ð¼Ð¾Ð´ÐµÐ»Ð¸
automl = autosklearn.classification.AutoSklearnClassifier(
    time_left_for_this_task=120, 
    per_run_time_limit=30
)
automl.fit(X_train, y_train)
```

```python
# ÐŸÑ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð¸Ñ Ð¸ Ð¾Ñ†ÐµÐ½ÐºÐ°
y_pred = automl.predict(X_test)
print("Auto-sklearn Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Ð›ÑƒÑ‡ÑˆÐ¸Ðµ Ð¼Ð¾Ð´ÐµÐ»Ð¸
print(automl.leaderboard())
```

```python
## âœ… AutoKeras: ÐšÐ»Ð°ÑÑÐ¸Ñ„Ð¸ÐºÐ°Ñ†Ð¸Ñ MNIST
```

```python
import autokeras as ak
from tensorflow.keras.datasets import mnist
```

```python
# Ð—Ð°Ð³Ñ€ÑƒÐ·ÐºÐ° MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# ÐŸÑ€Ð¸Ð²ÐµÐ´ÐµÐ½Ð¸Ðµ Ð´Ð°Ð½Ð½Ñ‹Ñ… Ðº Ñ„Ð¾Ñ€Ð¼Ð°Ñ‚Ñƒ (n, 28, 28, 1)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
```

```python
# ÐŸÐ¾ÑÑ‚Ñ€Ð¾ÐµÐ½Ð¸Ðµ AutoML Ð¼Ð¾Ð´ÐµÐ»Ð¸
clf = ak.ImageClassifier(overwrite=True, max_trials=2)
clf.fit(x_train, y_train, epochs=5)
```

```python
# ÐžÑ†ÐµÐ½ÐºÐ° Ð¼Ð¾Ð´ÐµÐ»Ð¸
_, acc = clf.evaluate(x_test, y_test)
print("AutoKeras Accuracy:", acc)
```

```python
## ðŸ§ª Ð¡Ñ€Ð°Ð²Ð½ÐµÐ½Ð¸Ðµ Ñ Ñ€ÑƒÑ‡Ð½Ð¾Ð¹ Ð¼Ð¾Ð´ÐµÐ»ÑŒÑŽ (Ð¾Ð¿Ñ†Ð¸Ð¾Ð½Ð°Ð»ÑŒÐ½Ð¾)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, rf_pred))
print(classification_report(y_test, rf_pred))
```
