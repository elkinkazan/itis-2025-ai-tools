```python
## ✅ Auto-sklearn: Классификация Breast Cancer
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import autosklearn.classification
```

```python
# Загрузка и разбиение данных
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Обучение AutoML модели
automl = autosklearn.classification.AutoSklearnClassifier(
    time_left_for_this_task=120, 
    per_run_time_limit=30
)
automl.fit(X_train, y_train)
```

```python
# Предсказания и оценка
y_pred = automl.predict(X_test)
print("Auto-sklearn Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Лучшие модели
print(automl.leaderboard())
```

```python
## ✅ AutoKeras: Классификация MNIST
```

```python
import autokeras as ak
from tensorflow.keras.datasets import mnist
```

```python
# Загрузка MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# Приведение данных к формату (n, 28, 28, 1)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
```

```python
# Построение AutoML модели
clf = ak.ImageClassifier(overwrite=True, max_trials=2)
clf.fit(x_train, y_train, epochs=5)
```

```python
# Оценка модели
_, acc = clf.evaluate(x_test, y_test)
print("AutoKeras Accuracy:", acc)
```

```python
## 🧪 Сравнение с ручной моделью (опционально)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, rf_pred))
print(classification_report(y_test, rf_pred))
```
