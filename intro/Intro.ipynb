```python
df.groupby('class')['mark'].mean()
```

```python
subset =df.groupby('class')['mark'].mean().reset_index()
subset.head()
```

```python
plt.figure(figsize=(10,6))
plt.plot(subset['class'], subset['mark'], label='Mark per class')
plt.xlabel('Mark')
plt.ylabel('Class')
plt.title('Mark per class')
plt.legend()
plt.show()
```
